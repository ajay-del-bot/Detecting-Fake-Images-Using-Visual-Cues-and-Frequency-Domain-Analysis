```python
import os
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap as lsc
import torch
import h5py

from deeplsd.utils.tensor import batch_to_device
from deeplsd.models.deeplsd_inference import DeepLSD
from deeplsd.geometry.viz_2d import plot_images, plot_lines



image_directory = '/content/drive/MyDrive/Fall 2024/CS 543/project/Recent_Kandinsky_Indoor/test/gen'
output_directory = '/content/drive/MyDrive/Fall 2024/CS 543/project/Recent_Kandinsky_Indoor/line_images_fake'

os.makedirs(output_directory, exist_ok=True)



device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
conf = {
    'detect_lines': True,
    'line_detection_params': {
        'merge': False,
        'filtering': 'strict',
        'grad_thresh': 3,
        'grad_nfa': True,
    }
}

ckpt = 'weights/deeplsd_md.tar'
ckpt = torch.load(str(ckpt), map_location='cpu')
net = DeepLSD(conf)
net.load_state_dict(ckpt['model'])
net = net.to(device).eval()

for img_file in image_files:
    img_path = os.path.join(image_directory, img_file)

    img = cv2.imread(img_path)[:, :, ::-1]
    gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    inputs = {'image': torch.tensor(gray_img, dtype=torch.float, device=device)[None, None] / 255.}
    with torch.no_grad():
        out = net(inputs)
        pred_lines = out['lines'][0]

    plt.figure(figsize=(10, 10))
    plot_images([img], [f'DeepLSD lines for {img_file}'], cmaps='gray')
    plot_lines([pred_lines], indices=range(1))
    plt.savefig(os.path.join(output_directory, f'lines_{img_file}'))
    plt.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from math import atan2, degrees, pi

def analyze_real_vs_fake_lines(real_folder, fake_folder, output_folder):
    """
    Compare line segment counts, lengths, and orientations for real vs fake images.
    - real_folder: Folder containing real images.
    - fake_folder: Folder containing generated images.
    - output_folder: Folder to save the results.
    """

    def process_images(folder):
        """Detect lines and return line counts, lengths, and orientations."""
        line_counts = []
        max_lengths = []
        orientations = []

        for img_name in os.listdir(folder):
            if img_name.endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(folder, img_name)
                img = cv2.imread(img_path, cv2.IMREAD_COLOR)
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

                edges = cv2.Canny(gray, 50, 150, apertureSize=3)
                lines = cv2.HoughLinesP(edges, 1, np.pi / 180, 50, minLineLength=30, maxLineGap=10)

                if lines is not None:
                    line_counts.append(len(lines))
                    lengths = [np.linalg.norm((x2-x1, y2-y1)) for x1, y1, x2, y2 in lines[:, 0]]
                    max_lengths.append(max(lengths) if lengths else 0)

                    angles = [atan2(y2 - y1, x2 - x1) for x1, y1, x2, y2 in lines[:, 0]]
                    orientations.extend(angles)
                else:
                    line_counts.append(0)
                    max_lengths.append(0)

        return line_counts, max_lengths, orientations

    def plot_histogram(data_real, data_fake, bins, title, xlabel, ylabel, output_name):
        """Utility to plot histograms."""
        plt.figure(figsize=(10, 6))
        plt.hist(data_real, bins=bins, alpha=0.6, label='Real', color='blue', edgecolor='black')
        plt.hist(data_fake, bins=bins, alpha=0.6, label='Generated', color='orange', edgecolor='black')
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.legend()
        plt.grid(True, linestyle=':')
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, output_name))
        plt.show()

    def plot_polar(orientations_real, orientations_fake, output_name):
        """Plot polar histograms for line orientations."""
        plt.figure(figsize=(12, 6))

        ax1 = plt.subplot(121, polar=True)
        ax1.hist(orientations_real, bins=36, alpha=0.7, color='black')
        ax1.set_title('Mean Line Orientation per Image (Real)')

        ax2 = plt.subplot(122, polar=True)
        ax2.hist(orientations_fake, bins=36, alpha=0.7, color='black')
        ax2.set_title('Mean Line Orientation per Image (Generated)')

        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, output_name))
        plt.show()

    real_counts, real_lengths, real_orientations = process_images(real_folder)
    fake_counts, fake_lengths, fake_orientations = process_images(fake_folder)

    plot_histogram(real_counts, fake_counts, bins=50,
                   title='Number of Lines (Real vs Generated Images)',
                   xlabel='Number of Lines',
                   ylabel='Number of Images',
                   output_name='line_counts.png')

    plot_histogram(real_lengths, fake_lengths, bins=50,
                   title='Maximum Line Segment Length (Real vs Generated Images)',
                   xlabel='Maximum Line Segment Length',
                   ylabel='Number of Images',
                   output_name='line_lengths.png')

    plot_polar(real_orientations, fake_orientations, output_name='line_orientations.png')

real_folder = "/content/drive/MyDrive/Fall 2024/CS 543/project/Recent_Kandinsky_Indoor/line_images_real"
fake_folder = "/content/drive/MyDrive/Fall 2024/CS 543/project/Recent_Kandinsky_Indoor/line_images_fake"
output_folder = "/content/drive/MyDrive/Fall 2024/CS 543/project"

os.makedirs(output_folder, exist_ok=True)
analyze_real_vs_fake_lines(real_folder, fake_folder, output_folder)
```
